# pandas practice 퀴즈 13문제

```python
import pandas as pd
data = { 'Date': ['2025-10-01', '2025-10-05', '2025-10-06', '2025-10-06', '2025-10-07'], 'Account': ['판관비', '매출', '매출', '판관비', '매출원가'], 'Amount': [50000, 2000000, 3500000, 1200000, 1500000] }
df_trans = pd.DataFrame(data)
df_trans['Date'] = pd.to_datetime(df_trans['Date'])
print("--- 원본 데이터 ---")
df_trans
```

1. [Pandas 필터링 & 시계열] "주말(토=5, 일=6)에 발생"하고 "금액(Amount)이 100만 원 초과"인 '고위험 의심 거래'를 필터링하는 코드를 작성

```python
df_trans['isweekend'] = df_trans['Date'].dt.dayofweek >= 5
df_trans['isamount'] = df_trans['Amount'] > 1000000
cond1 = df_trans['isweekend']
cond2 = df_trans['isamount']
df_trans[cond1 & cond2]
```

2. [Pandas Groupby & 파이썬 기초: f-string]
groupby와 agg를 사용해 'Account'(계정)별 "거래 건수(count)"와 "평균 금액(mean)"을 계산하는 summary 데이터프레임을 작성

```python
summary = df_trans.groupby('Account')['Amount'].agg(['count','mean'])
for account_name, row_data in summary.iterrows():
  print(f"계정[{account_name}] - 거래 건수:{row_data['count']}, 평균 금액:{row_data['mean']}")
```

3.[Pandas Apply & 파이썬 기초: def 함수]
def로 '금액(Amount)'을 입력받아 리스크 등급을 반환하는 함수 categorize_risk를 작성.

300만 원 이상: "고위험"

100만 원 이상 300만 원 미만: "중위험"

100만 원 미만: "저위험"

그다음, .apply()를 사용해 'RiskLevel'이라는 새 컬럼에 이 함수를 적용하고 최종 df_trans를 출력

```python
def categorize_risk(amount):
  if amount >= 3000000:
    return '고위험'
  elif amount >= 1000000 and amount < 3000000:
    return '중위험'
  else:
    return '저위험'
df_trans['RiskLevel'] = df_trans['Amount'].apply(categorize_risk)
df_trans
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "메뉴":['아메리카노', '카페라떼', '에스프레소', '카페모카', '바닐라라떼'],
    "가격":[4500, 5000, 4000, 5900, 5300],
    "칼로리":[10, 110, np.nan, 210, np.nan],
    "원산지":['과테말라','브라질','과테말라', np.nan, np.nan]
})
df
```

1. '칼로리' 컬럼 결측치는 칼로리 데이터 중 최소값으로 결측치를 채우시오

```python
cal_min = df['칼로리'].min()
df = df['칼로리'].fillna(cal_min)
df
```

2. '원산지' 컬럼 결측치는 원두 데이터 중 최빈값으로 결측치를 채우시오

```python
won_mode = df['원산지'].mode()[0]
df['원산지'] = df['원산지'].fillna(won_mode)
df
```
