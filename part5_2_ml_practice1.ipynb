## 데이터 불러오기

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ceredit%20card/train.csv")
test =  pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ceredit%20card/test.csv")
# train = pd.read_csv("train.csv")
# test =  pd.read_csv("test.csv")
```

## 데이터 전처리 & 피처엔지니어링

```python
cols = train.select_dtypes(include = "object").columns
cols
# #baseline (object 행 제거)
# train = train.drop(cols, axis=1)
# test = test.drop(cols, axis=1)

#label
from sklearn.preprocessing import LabelEncoder

for col in cols:
  le = LabelEncoder()
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])

# one-hot
# train = pd.get_dummies(train, columns=cols)
# test = pd.get_dummies(train, columns=cols)
# train.head()
```

```python
train = train.drop('CLIENTNUM', axis=1)
test_id = test.pop('CLIENTNUM')
```

```python
test.head(1)
```

```python
test_id
```

## 검증 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train.drop('Attrition_Flag', axis=1), train['Attrition_Flag'], test_size=0.2, random_state=0)
```

```python
X_tr.head()
```

## 모델 & 평가

```python
#모델 불러오기
#학습
#예측
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 0)
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
```

```python
pred
```

## 예측 및 csv 제출

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 정확도
print(accuracy_score(y_val, pred))

# 정밀도
print(precision_score(y_val, pred))

# 재현율 (민감도)
print(recall_score(y_val, pred))

# F1
print(f1_score(y_val , pred))

#baseline
# 0.9617520049352252
# 0.9147982062780269
# 0.8259109311740891
# 0.8680851063829788

#label
# 0.9629858112276373
# 0.9348837209302325
# 0.8137651821862348
# 0.8701298701298701

#one-hot
# 0.958667489204195
# 0.9205607476635514
# 0.7975708502024291
# 0.8546637744034707
```

```python
# roc-auc
pred = model.predict_proba(X_val)
print(roc_auc_score(y_val, pred[:,1]))
#baseline
# 0.9881886863615202
#label
# 0.9867551815379901
#one-hot
# 0.9870866703204096
```
