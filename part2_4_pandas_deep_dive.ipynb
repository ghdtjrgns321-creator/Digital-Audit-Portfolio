## 문제1.(select_dtype + groupby + def + return)

df_quiz에서 데이터 타입(dtype)이 'object'인 컬럼 (문자열 컬럼)만 선택하여 df_text라는 새 변수에 저장하세요.
def를 사용하여, 'Amount' 컬럼의 값이 '양수'면 'Income', '음수'면 'Expense'를 '반환(return)'하는 check_flow 함수를 만드세요.
df_quiz를 'Account'별로 그룹화(groupby)하여, 'Amount'의 '평균'을 구하는데, 방금 만든 check_flow 함수를 apply하여 'Income'/'Expense' 분류도 함께 집계하세요.

```python
import pandas as pd
import numpy as np
data = { 'InvoiceID': ['A1001', 'A1002', 'B1003', 'C1004', 'C1005', 'D1006'], 'Date': ['2025-10-01', '2025-10-02', '2025-10-02', '2025-10-03', '2025-10-30', '2025-11-01'], 'Account': ['매출', '매출원가', '판관비', '매출', '영업외비용', '판관비'], 'Amount': [5000000, -1500000, -800000, 7000000, -50000, -200000], 'Team': ['Sales_A', 'Cost_A', 'Ops_B', 'Sales_A', 'Finance', 'Ops_A'] }
df_quiz = pd.DataFrame(data)
df_quiz['Amount'] = df_quiz['Amount'].astype('float64')
df_quiz['Date'] = pd.to_datetime(df_quiz['Date'])
print("--- 퀴즈 원본 데이터 (df_quiz) ---")
display(df_quiz)

#df_quiz.info()
df_text = df_quiz.select_dtypes(include = 'object')
#df_text.info()
def check_flow(data):
  if data.mean() > 0:
    return 'Income'
  else:
    return 'Expense'
summary = df_quiz.groupby('Account')['Amount'].agg(['mean', check_flow])
summary.columns = ['Amount(Mean)', 'Flow']
display(summary)

```

## 문제3 (set_index + .loc + def + .apply)

df_quiz의 'Date' 컬럼을 새로운 '인덱스'로 설정하세요. (set_index 활용)
'2025-10-02' 하루 동안의 모든 거래 데이터만 .loc을 사용해 뽑아내세요.
def를 사용하여, 'Team' 컬럼의 값(예: 'Sales_A')을 '입구(파라미터)'로 받는 split_team 함수를 만드세요.
이 함수는 'Sales_A'처럼 _가 있으면, _ 뒤의 'A'만 '반환(return)'해야 합니다.
만약 'Finance'처럼 _가 없으면, 'N/A'를 '반환(return)'해야 합니다.
df_quiz의 'Team' 컬럼에 이 split_team 함수를 .apply()하여 'Team_Code'라는 새 컬럼을 추가하고, 최종 df_quiz를 출력하세요.

```python
import pandas as pd
import numpy as np
data = { 'InvoiceID': ['A1001', 'A1002', 'B1003', 'C1004', 'C1005', 'D1006'], 'Date': ['2025-10-01', '2025-10-02', '2025-10-02', '2025-10-03', '2025-10-30', '2025-11-01'], 'Account': ['매출', '매출원가', '판관비', '매출', '영업외비용', '판관비'], 'Amount': [5000000, -1500000, -800000, 7000000, -50000, -200000], 'Team': ['Sales_A', 'Cost_A', 'Ops_B', 'Sales_A', 'Finance', 'Ops_A'] }
df_quiz = pd.DataFrame(data)
df_quiz['Amount'] = df_quiz['Amount'].astype('float64')
df_quiz['Date'] = pd.to_datetime(df_quiz['Date'])
print("--- 퀴즈 원본 데이터 (df_quiz) ---")
display(df_quiz)

df_quiz = df_quiz.set_index('Date')
df_quiz.loc['2025-10-02']

def split_team(value):
  if "_" in value:
    return value.split("_")[1]
  else:
    return 'N/A'
df_quiz['Team_Code'] = df_quiz['Team'].apply(split_team)
df_quiz
```
